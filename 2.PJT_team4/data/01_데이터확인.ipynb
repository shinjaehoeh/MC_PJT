- 가산율 : 진료비 가산율이란 의료기관 종류별로 시설·인력·장비·서비스 수준의 차이로 인해 처치·검사·치료 등의 비용차이가 발생하기 때문에, 의료기관 종류별로 진료비 등 의료서비스 비용에 일정률을 더하는 것을 말합니다.
